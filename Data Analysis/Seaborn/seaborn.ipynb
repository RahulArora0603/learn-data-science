```python
import numpy as np
import pandas as pd 
import seaborn as sns # import seaborn
import matplotlib.pyplot as plt 
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
# 1. Pairplot - (relationships between variables)
sns.pairplot(iris, hue='species', palette='Set2') # hue decides the color
plt.suptitle("Pair Plot of Iris Dataset", y=1.02) # title
plt.show()
```

```python
# 2. Boxplot in Seaborn (better than plt)
sns.boxplot(data=iris, x='species', y='sepal_length')
plt.title("Sepal Length by Species")
plt.show()
```

```python
# 3. Violin Plot
sns.violinplot(data=iris, x='species', y='petal_length')
plt.title("Petal Length Distribution by Species")
plt.show()
```

```python
# 4. Heatmap - intensity based on value
corr = iris.drop(columns='species').corr() # correlation b/w numeric variables
sns.heatmap(corr, annot=True, cmap='coolwarm') # annot - shows values, cmap - colormap
plt.title("Correlation Matrix - Iris Dataset")
plt.show()
```

```python
# 5. Swarmplot 
sns.swarmplot(data=iris, x='species', y='petal_width', palette='muted')
plt.title("Petal Width by Species")
plt.show()
```

```python
# 6. Histplot
sns.histplot(data=iris, x='petal_length',hue='species', bins=20, kde=True, color='green') # 
plt.title("Distribution of Petal Length")
plt.show()
```

```python
# Use FacetGrid to plot subplots based on categories
g = sns.FacetGrid(iris, col="species")
g.map(sns.histplot, "petal_length")
plt.tight_layout()
plt.show()
```
