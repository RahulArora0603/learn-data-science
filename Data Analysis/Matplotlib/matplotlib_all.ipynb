```python
import pandas as pd
import matplotlib.pyplot as plt #used for plotting visuals
```

```python
import numpy as np
```

```python
# Basic Visualizations in Matplotlib
# 1. Bar Chart
categories = ['A','B','C','D','E']
values = [22,34,28,25,34]
plt.bar(categories, #categories
        values,     # value corresponding each category
        width=0.6,  # width of the bars
        color = ['red','blue','green','orange','teal']) # colors of the bars
plt.show() # to display the bar on screen 
```

```python
# 2. Pie Chart - categorical data
categories = ['A','B','C','D','E']
values = [22,34,28,25,34]
plt.pie(values, # the numeric values
        labels=categories, # the category names
        colors=['blue','red','yellow','orange','purple'], # color of each slice
        startangle=100, # rotates the chart by an angle
        autopct="%.2f", # percent of pie chart area a slice is covering
        shadow=True, # shadow of the chart
        radius=0.8) # radius of the chart
        # wedgeprops={'width':0.4}) - use this code to convert pie into doughnut chart
plt.show()
```

```python
# 3. Line Chart - for plotting time series data (sales trend over time)
categories = ['A','B','C','D','E']
values = [22,34,28,25,34]
plt.plot(categories, 
        values,
        color='green', 
        linestyle = '--', # line style
        linewidth = 1.5,  # width of line
        marker = 'o', # style of the marker (individual points)
        markersize=4, # size of marker
        markerfacecolor = 'red', # color of marker
        markeredgecolor='red')
plt.show()
```

```python
# 4. Scatter Plot - shows relation between two numeric columns 
list1 = [10,14,22,36,28,8,16,21,27]
list2 = [12,22,34,38,40,28,41,32,25]
plt.scatter(list1,
            list2,
            marker='o',
            color='black',
            edgecolor='green',
            alpha=1)
plt.show()
```

```python
# 5. Histogram - shows distribution of a numeric column
data = np.random.normal(loc=50, scale=10, size=1000)

plt.hist(data, # numeric data
        bins=20, # number of bins or groups
        color='skyblue', 
        edgecolor='black', 
        alpha=0.75, # transparency
        rwidth=0.9, # width of each bar
        histtype='bar') # type of histogram

plt.xlabel('Value') # x-axis title
plt.ylabel('Frequency') #y-axis title
plt.title('Histogram Example') # title
plt.legend() #legend
plt.grid(True)
plt.show()
```

```python
# 6. BoxPlot 
np.random.seed(42)
data = [np.random.normal(loc=mu, scale=5, size=100) for mu in [50, 60, 70]]

plt.boxplot(data, patch_artist=True, notch=True, vert=True,
            labels=['Group 1', 'Group 2', 'Group 3'],
            boxprops=dict(facecolor='gold', color='black'), # box attributes
            medianprops=dict(color='red'), # median line
            whiskerprops=dict(color='black'), # whiskers
            capprops=dict(color='green'), # caps
            flierprops=dict(marker='o', markerfacecolor='blue', markersize=6, linestyle='none'))

plt.title('Boxplot')
plt.ylabel('Values')
plt.grid(True)
plt.show()
```

```python
# 7. Violin Plot
np.random.seed(0)
data1 = np.random.normal(70, 10, 100)
data2 = np.random.normal(65, 15, 100)
data3 = np.random.normal(80, 5, 100)

data = [data1, data2, data3]

# Create the violin plot
plt.violinplot(data, showmeans=True, showmedians=True, showextrema=True,
               widths=0.7, vert=True)

plt.xticks([1, 2, 3], ['Group A', 'Group B', 'Group C']) # x-axis ticks
plt.title('Violin Plot Example') # chart title
plt.ylabel('Values') # y-axis title
plt.grid(True)
plt.show()
```

```python
cols = orders_df.select_dtypes(include=['float64','int64']).columns.to_list()
len(cols)
```

```python
# Matplotlib Subplots
fig,ax = plt.subplots(2,3) # fig is the figure and ax refers to all the blank plots inside it
n = 0
for i in range(2): # rows
    for j in range(3): # columns
        ax[i,j].boxplot(orders_df[cols[n]], patch_artist=True)
        ax[i,j].set_title(f"Histogram of {cols[n]}")
        n = n + 1
plt.tight_layout() # so that plots dont overlap
plt.show()        
```
